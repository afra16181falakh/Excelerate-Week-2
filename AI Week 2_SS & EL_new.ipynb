# Exploratory Data Analysis (EDA)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
students = pd.read_csv('C:/Users/Public/Downloads/AI_Internship_Week2.csv')
students.head()
```

```python
# 1. Count the number of NaN values in each column
empty_per_column = students.isna().sum()

# 2. Filter columns that have at least one NaN
columns_with_empty = empty_per_column[empty_per_column > 0]

# 3. Total number of NaN values across the entire dataframe
total_empty = students.isna().sum().sum()

print(f"Columns with empty entries:\n{columns_with_empty}")
print(f"Total number of empty entries: {total_empty}")
```

# Graphical Analysis

###### Insight Generation and Hypothesis Development is a critical phase in data analysis that involves deriving meaningful observations from data. A hypothesis is basically an educated guess or prediction that we can test based on the data.

## Gender Distribution

```python
plt.figure(figsize=(7,5))
sns.countplot(x='Gender', data=students, palette='pastel')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Number of students')
plt.show()
```

###### Insights:

###### What It Is: 
A bar chart showing the count of students by gender. 

###### What It Tells Us: 
This chart helps us see if there are more male or female students in the program. If there’s a big difference, it might mean the program appeals more to one gender, or there could be some barrier keeping the other gender from joining. 


###### Why It Matters: 
A balanced gender representation can make for a more inclusive environment. If one gender is less represented, it may indicate a need to create a welcoming atmosphere or promotional content that speaks to that group’s interests.

###### Hypothesis:
"The program appeals more to one gender, possibly due to differences in marketing, content, or perceived relevance." If there’s a big difference in gender representation, it could be that the program is unintentionally more attractive to one gender. Testing this hypothesis might involve surveying students about why they joined or analyzing the language and imagery in program promotions to see if they cater to one gender over the other.

## Learner SignUp Days in a week:

```python
# Plotting the number of sign-ups in different days of a week
plt.figure(figsize=(9,5))
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
sns.countplot(x='SignUp Day of Week', data=students, order=day_order, palette='dark')
plt.title('SignUp Day of Week')
plt.xlabel('Days of Week')
plt.ylabel('Number of Students')
plt.show()
```

###### Insight:
This bar chart displays the number of sign-ups per weekday. Peaks on certain days (i.e. Thursday) might suggest trends in learner behavior or preference, which could be useful in timing promotional activities or campaigns to drive higher engagement.

###### Hypothesis:
"Students are more likely to sign up for the program on certain days of the week(i.e. Thursday), possibly because of routine schedules or social media activity." If we observe higher sign-ups on specific days(i.e. Thursday), it could be linked to when people are more available to make decisions or when they’re more active online. We could test this by experimenting with marketing on different days and tracking changes in sign-up rates.

## Scatter Plotting between Age of Learner & Time in Opportunity vs Engagement Score

```python
# Age of Learner x Time in Opportunity vs Engagement Score
plt.figure(figsize=(10, 6))
sns.scatterplot(x=students['Engagement Score'], y=students['Age of Learner x Time in Opportunity'], color='coral')
plt.title("Age of Learner x Time vs Engagement Score")
plt.xlabel("Engagement Score")
plt.ylabel("Age x Time in Opportunity")
plt.show()
```

###### Insight:

###### What It Is: 
A scatter plot that shows the relationship between a student’s engagement score (how active or involved they are) and a combined factor of their age and time spent in the program.

###### What It Tells Us: 
This plot can reveal whether certain ages or lengths of time in the program lead to higher engagement. If older students who’ve been in the program longer show higher engagement, it suggests that experience may lead to more commitment.

###### Why It Matters: 
If specific age groups or experience levels have high engagement, then tailoring support for other groups (like younger or newer students) could increase their involvement and satisfaction.

###### Hypothesis:
"Older students or those who have been in the program longer tend to engage more." 
If engagement scores are higher for older or more experienced students, it may mean they have more commitment, stability, or interest in the subject. Testing this could involve offering additional support to younger or newer students to see if it raises their engagement levels.

## Histogram Plot of Distribution of Days Since Last Engagement

```python
plt.figure(figsize=(10, 6))
sns.histplot(students['Days Since Last Engagement'], bins=20, kde=True, color="mediumpurple")
plt.title("Distribution of Days Since Last Engagement")
plt.xlabel("Days Since Last Engagement")
plt.ylabel("Frequency")
plt.show()
```

###### Insight:

###### What It Is: 
A histogram that shows how long it’s been since students last participated in the program.

###### What It Tells Us: 
This chart lets us see how recently students have been active. If most students were active within the last few days, it’s a good sign of high engagement. However, if many haven’t engaged in a long time, it might mean they’re losing interest.

###### Why It Matters: 
This insight can guide when and how to re-engage students who have been inactive. For instance, if many students haven’t engaged for weeks, sending reminders or incentives could bring them back.

###### Hypothesis:
"Students who have been inactive for a long period are more likely to drop out." A long gap since the last engagement might signal declining interest or struggles with the program content. This hypothesis could be tested by sending re-engagement emails to inactive students and observing if it reduces dropout rates.

## Correlation Heatmap for numerical features

```python
plt.figure(figsize=(12, 10))
correlation_matrix = students.select_dtypes(include=[np.number]).corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', square=True, linewidths=0.5)
plt.title("Heatmap of Correlations Between Numerical Features")
plt.show()
```

###### Insight:

###### What It Is: 
A grid showing how different numerical features relate to each other. High numbers mean a strong relationship between two features. 

###### What It Tells Us: 
The heatmap can reveal which factors influence each other. For instance, if “Engagement Score” is highly related to another feature, that feature may play a key role in keeping students engaged. 

###### Why It Matters: 
Strong correlations help identify which factors are most influential. For example, if “time in the program” and “engagement” are closely related, it suggests that helping students stay longer could boost their engagement scores.

###### Hypothesis:
"Certain factors, like the time in the program, have a strong influence on engagement scores." If the heatmap shows high correlations, it suggests these factors might drive engagement. Testing could involve tracking students with different time lengths in the program and seeing if extended time directly boosts their engagement.

## Box Plot of Engagement Score by Opportunity Category

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Opportunity Category', y='Engagement Score', data=students, palette="pastel")
plt.title("Engagement Score by Opportunity Category")
plt.xlabel("Opportunity Category")
plt.ylabel("Engagement Score")
plt.xticks(rotation=45)
plt.show()
```

###### Insight:

###### What It Is: 
A box plot that shows the spread of engagement scores within different categories of opportunities (like courses, seminars, or certifications).

######  What It Tells Us: 
This plot shows if some types of opportunities are more engaging than others. For example, if seminars have a wider range of high engagement scores, students might find them more interesting or valuable.

###### Why It Matters: 
Knowing which categories have higher engagement can guide program improvements. If courses have lower engagement, perhaps more interactive elements could make them more appealing.

###### Hypothesis: 
"Some types of opportunities (like seminars or courses) are more engaging than others." If certain opportunity categories show higher engagement scores, it could mean students find them more interesting or valuable. Testing this might involve surveying students on what they enjoy most or adding similar opportunities to see if engagement rises.

## Pie-Chart of Current Student Status

```python
data = {
    'Current Student Status': ['Enrolled', 'Graduated', 'Enrolled', 'Dropped Out', 'Enrolled', 'Graduated', 'Enrolled']
}
students = pd.DataFrame(data)

# Count the occurrences of each status
status_counts = students['Current Student Status'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(status_counts, labels=status_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.tab10.colors)
plt.title('Current Student Status Distribution')
plt.axis('equal')
plt.show()
```

###### Insight:

###### What It Is:
A pie chart that shows the percentage of students in each status category, such as “Enrolled,” “Graduated,” or “Dropped Out.” 

###### What It Tells Us: 
This chart helps us understand what percentage of students are still enrolled, have graduated, or have dropped out. A high dropout rate might mean there are issues in the program, while a high graduation rate indicates success. 

###### Why It Matters: 
If many students are dropping out, understanding their reasons could lead to adjustments in the program. High graduation rates can be celebrated and used as a selling point in marketing.

###### Hypothesis: 
"Students with certain statuses (like ‘Enrolled’) are more likely to complete the program than those with other statuses (like ‘Started’). " If students in certain status categories have better outcomes, it could indicate that these stages support commitment. We could test this by offering additional guidance or incentives to students early on to help them reach the more stable "Enrolled" status.

## Pie Chart of Status Description

```python
data = {
    'Status Description': ['Started', 'Team Allocated', 'Withdraw']
}
students = pd.DataFrame(data)

status_counts = students['Status Description'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(status_counts, labels=status_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Set2.colors)
plt.title('Distribution of Status Description')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is circular.
plt.show()
```

###### Insight: 


###### What It Is: 
Another pie chart that shows the percentage of students in different stages, like “Started,” “Team Allocated,” or “Withdrawn.” 

###### What It Tells Us: 
This pie chart shows where students are in their journey. High numbers in “Withdrawn” might indicate obstacles in early stages, while many in “Team Allocated” may show smooth progress. 

###### Why It Matters: 
If many students are stuck or dropping out at certain stages, program administrators can investigate those stages for improvements. For example, if “Started” has the most students but many aren’t moving forward, additional orientation resources may be helpful.

###### Hypothesis: 
"Students who reach advanced stages in the program are less likely to drop out." If the dropout rate decreases as students progress, it might mean they feel more committed as they get closer to completing the program. Testing could involve providing extra support at early stages to encourage progression and reduce dropouts.
