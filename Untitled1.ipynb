```python
# import necessary data
import pandas as pd 
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as py
import matplotlib.pyplot as plt
```

```python

import os

# Define the file path
file_path = 'C:/users/LENOVO/Downloads/Updated_Cleaned_Data.csv'

# Check if the file exists before trying to read it
if os.path.exists(file_path):
    print("File exists. Proceeding to load the data...")
    # Load the CSV file into a DataFrame
    cleaned_data = pd.read_csv(file_path)
    # Display the first few rows to confirm it's loaded correctly
    print(cleaned_data.head())
else:
    print("File does not exist. Please check the file path.")

```

```python
# Gender distribution
plt.figure(figsize=(6,4))
sns.countplot(x='Gender' ,data=students, palette='cool')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Enagament over time

# Convert 'Apply Date' to datetime format
cleaned_data['Apply Date'] = pd.to_datetime(cleaned_data['Apply Date'])

# Group by month to get average engagement over time
monthly_engagement = cleaned_data.resample('ME', on='Apply Date')['Engagement Score'].mean().reset_index()

# Line plot
sns.lineplot(x='Apply Date', y='Engagement Score', data=monthly_engagement)
plt.title('Average Engagement Over Time')
plt.xlabel('Month')
plt.ylabel('Average Engagement Score')
plt.xticks(rotation=45)
plt.show()
```

```python
# 2. Age vs. Engagement
sns.scatterplot(x='Age of Learner', y='Engagement Score', data=cleaned_data)
plt.title('Age of Learner vs. Engagement Score')
plt.xlabel('Age of Learner')
plt.ylabel('Engagement Score')
plt.show()
```

```python
# 3. Opportunity Category vs. Engagement
sns.barplot(x='Opportunity Category', y='Engagement Score', data=cleaned_data)
plt.title('Average Engagement Score by Opportunity Category')
plt.xticks(rotation=45)
plt.show()
```

```python
# 4. Sign-Up Time Analysis
heatmap_data = cleaned_data.pivot_table(index='SignUp Day of Week', columns='SignUp Month', values='Profile Id', aggfunc='count')
sns.heatmap(heatmap_data, cmap='Blues', annot=True)
plt.title('Sign-Ups Heatmap: Day of Week vs. Month')
plt.xlabel('SignUp Month')
plt.ylabel('Day of the Week')
plt.show()
```

```python
# 5. Duration in Opportunity vs. High Engagement
sns.boxplot(x='High Engagement', y='Time in Opportunity (Days)', data=cleaned_data)
plt.title('Time in Opportunity vs. Engagement Level')
plt.xlabel('High Engagement')
plt.ylabel('Time in Opportunity (Days)')
plt.show()
```

```python
# 6. Days Since Last Engagement vs. Re-Engagement
sns.histplot(data=cleaned_data, x='Days Since Last Engagement', hue='High Engagement', multiple='stack')
plt.title('Days Since Last Engagement for High vs. Low Engagement')
plt.xlabel('Days Since Last Engagement')
plt.ylabel('Count')
plt.show()
```

```python
# 7. Major vs. Engagement
sns.boxplot(x='Current/Intended Major', y='Engagement Score', data=cleaned_data)
plt.title('Engagement Score by Major')
plt.xticks(rotation=45)
plt.show()
```

```python
# 8. Status Code vs. Time in Opportunity
sns.barplot(x='Status Code', y='Time in Opportunity (Days)', data=cleaned_data)
plt.title('Time in Opportunity by Status Code')
plt.xlabel('Status Code')
plt.ylabel('Time in Opportunity (Days)')
plt.show()
```

```python
# 9. Geographical Analysis
sns.barplot(x='Country', y='Engagement Score', data=cleaned_data)
plt.title('Average Engagement Score by Country')
plt.xticks(rotation=45)
plt.show()
```

```python
# 10. Monthly Engagement Trend
# Ensure the 'Learner SignUp DateTime' column is in datetime format
cleaned_data['Learner SignUp DateTime'] = pd.to_datetime(cleaned_data['Learner SignUp DateTime'], errors='coerce')

# Extract Month-Year period for plotting
cleaned_data['Signup_MonthYear'] = cleaned_data['Learner SignUp DateTime'].dt.to_period('M')

# Group by 'Signup_MonthYear' and calculate average engagement score
monthly_engagement = cleaned_data.groupby('Signup_MonthYear')['Engagement Score'].mean().reset_index()

# Convert 'Signup_MonthYear' to string for plotting
monthly_engagement['Signup_MonthYear'] = monthly_engagement['Signup_MonthYear'].astype(str)

# Plotting
sns.lineplot(x='Signup_MonthYear', y='Engagement Score', data=monthly_engagement)
plt.xticks(rotation=45)
plt.title('Monthly Average Engagement Score')
plt.xlabel('Month-Year')
plt.ylabel('Average Engagement Score')
plt.show()
```
